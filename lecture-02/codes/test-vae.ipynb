```python
from models import VanillaVAE

import torch
from torchinfo import summary
```

```python
def test_vanilla_vae():
    # Instantiate model
    model = VanillaVAE(3, 10)

    print("=== Model Summary ===")
    print(summary(model, input_size=(1, 3, 64, 64)))

    # Forward pass test
    print("\n=== Forward Test ===")
    x = torch.randn(16, 3, 64, 64)
    out = model(x)
    print("Output[0] shape:", out[0].shape)

    # Loss test
    print("\n=== Loss Test ===")
    loss = model.loss_function(*out, M_N=0.005)
    print("Loss:", loss)
```

```python
test_vanilla_vae()
```
